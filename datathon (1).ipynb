```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
df = pd.read_csv('cleaned_dataset.csv')
```

```python
df
```

```python
df = df.rename(columns = {"Q5.Feelings":"Q3.Feelings","Q7.Utility_grade":"Q4.Utility_grade","Q8.Advantage_teaching":"Q5.Advantage_teaching","Q9.Advantage_learning":"Q6.Advantage_learning","Q10.Advantage_evaluation":"Q7.Advantage_evaluation","Q11.Disadvantage_educational_process":"Q8.Disadvantage_educational_process","Q12.Gender":"Q9.Gender","Q13.Year_of_study":"Q10.Year_of_study","Q14.Major":"Q11.Major","Q16.GPA":"Q12.GPA"})
```

```python
df.columns
```

```python
sns.set_style("whitegrid")

# Visualization for Perceived Usefulness of AI in Education (Q4.Utility_grade)
plt.figure(figsize=(8,6))
sns.histplot(df['Q4.Utility_grade'], kde=False, bins=7, color='skyblue')
plt.title('Perceived Usefulness of AI in Education')
plt.xlabel('Usefulness Rating (1-10)')
plt.ylabel('Number of Students')
plt.show()

```

```python
numerical_data = df[['Q1.AI_knowledge', 'Q4.Utility_grade', 'Q9.Gender', 'Q10.Year_of_study', 'Q12.GPA']]
correlation_matrix = numerical_data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Selected Variables')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Q4.Utility_grade', data=df, palette='viridis')
plt.title('Utility of AI in the Educational Process')
plt.xlabel('Utility Grade (1-10)')
plt.ylabel('Count')
plt.show()

# Function to plot the distribution of advantages/disadvantages
def plot_distribution(df, column, title, labels):
    plt.figure(figsize=(8, 5))
    sns.countplot(x=column, data=df, palette='coolwarm')
    plt.title(title)
    plt.xlabel('Options')
    plt.ylabel('Count')
    plt.xticks(range(len(labels)), labels, rotation=45)
    plt.show()

# Perceived Advantages in Teaching (Q8)
advantages_teaching_labels = ['Virtual Assistant', 'Time Management', 'Interactive Lessons', 'Other']
plot_distribution(df, 'Q5.Advantage_teaching', 'Main Advantage of AI in Teaching', advantages_teaching_labels)

# Perceived Advantages in Learning (Q9)
advantages_learning_labels = ['Personalized Lessons', 'Universal Access', 'Interactive Lessons', 'Other']
plot_distribution(df, 'Q6.Advantage_learning', 'Main Advantage of AI in Learning', advantages_learning_labels)

# Perceived Advantages in Evaluation (Q10)
advantages_evaluation_labels = ['Automation of Grading', 'Fewer Errors', 'Constant Feedback', 'Other']
plot_distribution(df, 'Q10.Advantage_evaluation ', 'Main Advantage of AI in Evaluation', advantages_evaluation_labels)

# Perceived Disadvantages (Q11)
disadvantages_labels = ['Lack of Relationship', 'Internet Addiction', 'Rarer Interactions', 'Loss of Information']
plot_distribution(df, 'Q8.Disadvantage_educational_process', 'Main Disadvantage of AI in Education', disadvantages_labels)
```

```python
numerical_data = df[['Q1.AI_knowledge', 'Q4.Utility_grade']]
correlation_matrix = numerical_data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Selected Variables')
plt.show()
```
